```python
label_cols=['HomePlanet','Destination','Cabin Deck','Cabin Side']
onehot_cols=['CryoSleep','VIP']
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder

le=LabelEncoder()
ohe=OneHotEncoder(sparse_output=False,drop='first')
```

```python
#label encoding
for col in label_cols:
  train[col]=le.fit_transform(train[col])
  test[col]=le.transform(test[col])
```

```python
#one hot encoding
for col in onehot_cols:
  encoded_train=ohe.fit_transform(train[[col]])
  encoded_train_df=pd.DataFrame(encoded_train,columns=ohe.get_feature_names_out([col]))
  train=pd.concat([train,encoded_train_df],axis=1)

  encoded_test=ohe.transform(test[[col]])
  encoded_test_df=pd.DataFrame(encoded_test,columns=ohe.get_feature_names_out([col]))
  test=pd.concat([test,encoded_test_df],axis=1)
```

```python
train_transport=ohe.fit_transform(train[['Transported']])
train_transport_df=pd.DataFrame(train_transport,columns=ohe.get_feature_names_out(['Transported']))
train=pd.concat([train,train_transport_df],axis=1)
```

```python
#dropping the onehot columns
train.drop(onehot_cols,axis=1,inplace=True)
```

```python
test.drop(onehot_cols,axis=1,inplace=True)
```

Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix
from sklearn.preprocessing import StandardScaler
```

```python
# Define features (X) and target (y) for training data
X = train.drop(['PassengerId', 'Cabin', 'Name', 'Transported','Transported_True'], axis=1)
y = train['Transported_True']

# Define features for test data, ensuring the same columns as X
x_test = test.drop(['PassengerId', 'Cabin', 'Name', 'Group'], axis=1)

# Initialize the StandardScaler
#scale = StandardScaler()

# Fit the scaler on the training features and transform both training and test features
#X_scale = scale.fit_transform(X)
#x_scale = scale.transform(x_test)
```

```python

# Split the scaled training data for model training and evaluation
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train the Random Forest Classifier
model = RandomForestClassifier(n_estimators=1000,criterion='gini', max_depth=15,random_state=42)
model.fit(X_train, y_train)

# Make predictions on the scaled Validation set
y_pred = model.predict(X_val)

# Evaluate the model
accuracy = accuracy_score(y_val, y_pred)
conf_matrix = confusion_matrix(y_val, y_pred)
class_report = classification_report(y_val, y_pred)

print(f"Accuracy: {accuracy}")
print("\nConfusion Matrix:")
print(conf_matrix)
print("\nClassification Report:")
print(class_report)
```

```python
import lightgbm as lgb

# Initialize and train the LightGBM Classifier
lgb_model = lgb.LGBMClassifier(random_state=42)
lgb_model.fit(X_train, y_train)

# Make predictions on the scaled Validation set
y_pred_lgb = lgb_model.predict(X_val)

# Evaluate the model
accuracy_lgb = accuracy_score(y_val, y_pred_lgb)
conf_matrix_lgb = confusion_matrix(y_val, y_pred_lgb)
class_report_lgb = classification_report(y_val, y_pred_lgb)

print(f"LightGBM Accuracy: {accuracy_lgb}")
print("\nLightGBM Confusion Matrix:")
print(conf_matrix_lgb)
print("\nLightGBM Classification Report:")
print(class_report_lgb)
```

# Task
Tune the hyperparameters of a Random Forest and a LightGBM model using grid search with cross-validation and compare their performance.

## Define the hyperparameter grid

### Subtask:
for both the Random Forest and LightGBM models.


**Reasoning**:
Define the hyperparameter grids for Random Forest and LightGBM models as dictionaries.



```python
param_grid_rf = {
    'n_estimators': [100, 200, 500],
    'max_depth': [10, 15, 20],
    'criterion': ['gini', 'entropy']
}

param_grid_lgb = {
    'n_estimators': [100, 200, 500],
    'max_depth': [10, 15, 20],
    'learning_rate': [0.01, 0.05, 0.1]
}
```

## Perform grid search with cross-validation

### Subtask:
Perform grid search with cross-validation on the training data for the Random Forest model using the defined parameter grid.


**Reasoning**:
Perform grid search with cross-validation on the training data for the Random Forest model using the defined parameter grid.



```python
from sklearn.model_selection import GridSearchCV

rf_model = RandomForestClassifier(random_state=42)

grid_search_rf = GridSearchCV(estimator=rf_model, param_grid=param_grid_rf, cv=5,n_jobs=-1,
                               scoring='accuracy')

grid_search_rf.fit(X_train, y_train)
```

### Subtask:
Perform grid search with cross-validation on the training data for the LightGBM model using the defined parameter grid.

**Reasoning**:
Perform grid search with cross-validation on the training data for the LightGBM model using the defined parameter grid.

```python
lgb_model = lgb.LGBMClassifier(random_state=42)

grid_search_lgb = GridSearchCV(estimator=lgb_model, param_grid=param_grid_lgb, cv=5,n_jobs=-1,
                                 scoring='accuracy')

grid_search_lgb.fit(X_train, y_train)
```

## Evaluate tuned models

### Subtask:
Train both models with their best parameters on the training set and evaluate their performance on the validation set using appropriate metrics (e.g., accuracy, confusion matrix, classification report).

```python
# Train the best Random Forest model and evaluate on the validation set
best_rf_model = grid_search_rf.best_estimator_
best_rf_model.fit(X_train, y_train)
y_pred_rf = best_rf_model.predict(X_val)

print("Random Forest Model Evaluation on Validation Set:")
print(f"Accuracy: {accuracy_score(y_val, y_pred_rf)}")
print("\nConfusion Matrix:")
print(confusion_matrix(y_val, y_pred_rf))
print("\nClassification Report:")
print(classification_report(y_val, y_pred_rf))

# Train the best LightGBM model and evaluate on the validation set
best_lgb_model = grid_search_lgb.best_estimator_
best_lgb_model.fit(X_train, y_train)
y_pred_lgb_tuned = best_lgb_model.predict(X_val)

print("\nLightGBM Model Evaluation on Validation Set:")
print(f"Accuracy: {accuracy_score(y_val, y_pred_lgb_tuned)}")
print("\nConfusion Matrix:")
print(confusion_matrix(y_val, y_pred_lgb_tuned))
print("\nClassification Report:")
print(classification_report(y_val, y_pred_lgb_tuned))
```
